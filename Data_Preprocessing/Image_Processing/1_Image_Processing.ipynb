```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

##Â Read a color image

```python
img = cv2.imread('galaxies.jpg',1)
```

```python
plt.imshow(img)
plt.show()
```

```python
img.shape
```

## Convert the image to gray-scale

```python
img_converted = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img_converted)
plt.show()
```

```python
img_converted.shape
```

##Â Read the image directly in gray-scale

```python
img_gray = cv2.imread('galaxies.jpg',0) # note the 0 instead of 1
```

```python
plt.imshow(img_gray)
plt.show()
```

##Â Save the image

```python
cv2.imwrite('gray_image.jpg', img_gray)# params=None
```

## Crop the image

```python
plt.imshow(img[0:200,0:200])
plt.show()
```

## Resize the image

```python
image_100x100 = cv2.resize(img_gray, (100, 100))
```

```python
plt.imshow(image_100x100)
plt.show()
```
